```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Set up the Chrome WebDriver path
CHROME_DRIVER_PATH = 'path_to_your_chromedriver'

# Set up Chrome options (optional: headless mode)
options = Options()
options.headless = False  # set to True for headless (no UI) mode

# Initialize the Chrome WebDriver
driver = webdriver.Chrome(service=Service(CHROME_DRIVER_PATH), options=options)

def login_to_website():
    """
    Automates logging into a website with provided credentials.
    """
    driver.get("https://www.example.com/login")  # Replace with your website URL

    # Find the email field and enter your email
    email_input = driver.find_element(By.ID, "email")  # Adjust based on the website's structure
    email_input.send_keys("your_email@example.com")

    # Find the password field and enter your password
    password_input = driver.find_element(By.ID, "password")  # Adjust based on the website's structure
    password_input.send_keys("your_password")
    password_input.send_keys(Keys.RETURN)  # Press Enter to submit the login form

    # Wait for login to complete
    time.sleep(3)

def search_and_interact():
    """
    Automates searching on the website and interacting with the results.
    """
    # Perform a search after logging in
    search_box = driver.find_element(By.NAME, "q")  # Adjust based on the website's structure
    search_box.send_keys("Selenium WebDriver")
    search_box.send_keys(Keys.RETURN)

    # Wait for search results to load
    time.sleep(3)

    # Click on the first search result
    first_result = driver.find_element(By.XPATH, "(//h3)[1]")  # Adjust based on the website's structure
    first_result.click()

    # Wait for the page to load
    time.sleep(3)

    # Example: Interact with a button on the page after clicking the search result
    button = driver.find_element(By.XPATH, "//button[@id='action-button']")  # Adjust based on the website's structure
    button.click()

    # Wait to observe the interaction
    time.sleep(3)

def main():
    try:
        login_to_website()
        search_and_interact()
        print("Automation task completed successfully.")
    finally:
        driver.quit()

if __name__ == "__main__":
    main()
```
